```python
import pandas as pd
import numpy as np

# Load the CSV file
file_path = '/content/top_data (1).csv'
data = pd.read_csv(file_path)

# Generate random CPU usage between 0% and 100%
data['%CPU'] = np.random.uniform(low=0, high=100, size=len(data))

# Generate random memory usage as minutes:seconds (similar format)
minutes = np.random.randint(low=0, high=60, size=len(data))
seconds = np.random.randint(low=0, high=60, size=len(data))
data['MEM'] = ['{:02d}:{:02d}'.format(minute, second) for minute, second in zip(minutes, seconds)]

# Save the updated CSV file with random CPU and memory usage
output_file_path = 'top_data_with_usage.csv'
data.to_csv(output_file_path, index=False)

print(f"Updated file saved as {output_file_path}")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Visualizing a distribution
sns.histplot(df['%CPU'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Visualizing a distribution
sns.histplot(df['MEM'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'Timestamp' is in datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Plot CPU and Memory Usage over time
plt.figure(figsize=(14, 7))
plt.plot(df['Timestamp'], df['%CPU'], label='%CPU', color='blue')
plt.plot(df['Timestamp'], df['MEM'], label='Memory', color='green')

plt.title('CPU and Memory Usage Over Time')
plt.xlabel('Time')
plt.ylabel('Usage')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Sort by %CPU and plot top 10 processes
top_cpu = df.sort_values(by='%CPU', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_cpu['COMMAND'], top_cpu['%CPU'], color='orange')
plt.title('Top 10 Processes Consuming CPU')
plt.xlabel('% CPU Usage')
plt.ylabel('Process')
plt.gca().invert_yaxis()  # Invert y-axis to have the top process at the top
plt.show()
```

```python
# Sort by Memory and plot top 10 processes
top_mem = df.sort_values(by='MEM', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_mem['COMMAND'], top_mem['MEM'], color='purple')
plt.title('Top 10 Processes Consuming Memory')
plt.xlabel('Memory Usage')
plt.ylabel('Process')
plt.gca().invert_yaxis()
plt.show()
```
